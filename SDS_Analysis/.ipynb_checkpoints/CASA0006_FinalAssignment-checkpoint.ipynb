### 1.0 Introduction
The purpose of this paper is to examine factors that significantly contribute to energy consumption in the UK and investigate the segment the UK consumers based on energy usage. Following economic sanctions placed on Russian oil in 2022, energy prices in Europe increased dramatically. This led to public discontent and *knee-jerk* relief payments provided by the public sector. While these  payments provided short-term relief, investigating the nature of energy consumption in UK and the factors contributing to it can inform more targetted policy which could make the UK more reslient against price shocks and investigate factors that contribute to high energy consumption.

With this in mind, we seek to investgate 2 main research questions.

            RQ1 | Can we use spatial clustering to segment the UK population based on electricity and gas consumption?
            RQ2 | What are the main factors contributing to the energy consumption of each segment?

This study will leverage open data from the *Office of National Statistics* from **2022** and energy consumption data from the *UK Department for Energy Security and Net Zero*.

#### 5.1 Segmentation of UK Energy Consumers: Clustering

```python
# read in consumption data
gdf = gpd.read_file('ukge_consumption.gpkg')
gdf.head(1)
```

```python
# conduct spatial clusterin  using kmeans
# extract x, y values from geometry column
gdf['x'] = gdf.geometry.x
gdf['y'] = gdf.geometry.y

# prepare data for clustering & initialise kmeans
X = gdf[['ngas_median_cons_kwh', 'nelec_median_cons_kwh', 'x', 'y']]

# initialise HDBSCAN
# min_cluster_size = 5
clusterer = hdbscan.HDBSCAN(min_cluster_size=5, gen_min_span_tree=True)

# fit model
cluster_labels = clusterer.fit_predict(X)

# assign the cluster labels tp gdf
gdf['cluster'] = cluster_labels

# plot
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(column='cluster', categorical=True, legend=True, ax=ax)
ax.set_title('HDBSCAN Clustering of Energy Consumption')
plt.show()
```

```python
# produce information about each cluster
cluster_info = gdf.groupby('cluster').agg({
    'ngas_median_cons_kwh': ['mean', 'median'],
    'nelec_median_cons_kwh': ['mean', 'median']}).reset_index()

cluster_info.columns = ['Cluster', 'Gas Consumption Mean', 'Gas Consumption Median',
                        'Electricity Consumption Mean', 'Electricity Consumption Median']

# print
cluster_info
```
